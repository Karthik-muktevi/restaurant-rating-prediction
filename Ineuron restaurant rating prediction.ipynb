DATA PREPROCESSING

```python
df.head()
```

```python
df.drop(['url','address','name','phone','reviews_list','menu_item','listed_in(city)','dish_liked'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(), inplace = True)
```

```python
df['location'].fillna(df['rate'].mode()[0],inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
map_ = {'Yes':1,'No':0}
```

```python
df['online_order'] = df['online_order'].map(map_)
```

```python
df['book_table'] = df['book_table'].map(map_)
```

```python
df['listed_in(type)'].unique()
```

```python
type_={'Buffet':1, 'Cafes':2, 'Delivery':3, 'Desserts':4, 'Dine-out':5,'Drinks & nightlife':6, 'Pubs and bars':7}
df['listed_in(type)']=df['listed_in(type)'].map(type_)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df.location = le.fit_transform(df.location)
df.rest_type = le.fit_transform(df.rest_type)
df.cuisines = le.fit_transform(df.cuisines)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('rate', axis=1)
y = df['rate']
```

```python
x_train, x_test,y_train, y_test = train_test_split(x,y, test_size=0.20)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
dt = DecisionTreeRegressor()
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred = dt.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
rf = RandomForestRegressor()
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred1 = rf.predict(x_test)
```

```python
r2_score(y_test,y_pred1)
```

```python
import xgboost
from xgboost import XGBRegressor
```

```python
xg  =XGBRegressor()
```

```python
xg.fit(x_train,y_train)
```

```python
y_pred2 = xg.predict(x_test)
```

```python
r2_score(y_test,y_pred2)
```
